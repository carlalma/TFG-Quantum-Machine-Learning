```python
#IMPORTACIONES Y RUTAS
from pathlib import Path

import pandas as pd


# Permite ejecutar el notebook desde la carpeta notebooks
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()

BREAST_CANCER_PATH = PROJECT_ROOT / "data" / "raw" / "breast_cancer.csv"
WINE_QUALITY_PATH = PROJECT_ROOT / "data" / "raw" / "wine_quality.csv"

if not BREAST_CANCER_PATH.exists():
    raise FileNotFoundError(f"No se encuentra: {BREAST_CANCER_PATH}")

if not WINE_QUALITY_PATH.exists():
    raise FileNotFoundError(f"No se encuentra: {WINE_QUALITY_PATH}")
```

```python
#CARGAMOS LOS ARCHIVOS
breast_cancer_df = pd.read_csv(BREAST_CANCER_PATH)
wine_quality_df = pd.read_csv(WINE_QUALITY_PATH)

print("Breast Cancer:", breast_cancer_df.shape)
print("Wine Quality:", wine_quality_df.shape)
```

```python
#VISUALIZAMOS LA ESTRUCTURA DE LOS DATOS
display(breast_cancer_df.head())
display(wine_quality_df.head())

print("Columnas de Breast Cancer:")
print(breast_cancer_df.columns.tolist())

print("\nColumnas de Wine Quality:")
print(wine_quality_df.columns.tolist())
```

```python
#FUNCIÓN DE ANÁLISIS GENERAL
def summarize_dataset(
    dataframe: pd.DataFrame,
    name: str,
    target_column: str,
) -> None:
    """Muestra la información básica necesaria para validar un dataset."""

    print("=" * 70)
    print(name)
    print("=" * 70)

    print(f"Muestras: {dataframe.shape[0]}")
    print(f"Columnas: {dataframe.shape[1]}")
    print(f"Duplicados: {dataframe.duplicated().sum()}")
    print(f"Valores ausentes totales: {dataframe.isna().sum().sum()}")

    print("\nTipos de datos:")
    print(dataframe.dtypes.value_counts())

    print("\nValores ausentes por columna:")
    missing_values = dataframe.isna().sum()
    print(missing_values[missing_values > 0])

    print("\nDistribución de la variable objetivo:")
    print(dataframe[target_column].value_counts().sort_index())

    print("\nDistribución porcentual:")
    print(
        dataframe[target_column]
        .value_counts(normalize=True)
        .sort_index()
        .mul(100)
        .round(2)
    )
```

```python
#ANALIZAMOD AMBOS DATASETS
summarize_dataset(
    dataframe=breast_cancer_df,
    name="Breast Cancer Wisconsin",
    target_column="diagnosis",
)

summarize_dataset(
    dataframe=wine_quality_df,
    name="Wine Quality",
    target_column="quality",
)
```

```python
#VEMOS LAS COLUYMNAS QUE NO QUEREMOS

#Breast
breast_columns_to_remove = [
    column
    for column in ["id", "Unnamed: 32"]
    if column in breast_cancer_df.columns
]

print("Columnas que se eliminarán:", breast_columns_to_remove)

#Wine
wine_columns_to_remove = [
    column
    for column in ["Id", "id"]
    if column in wine_quality_df.columns
]

print("Columnas que se eliminarán:", wine_columns_to_remove)
```

```python
#DISENO DE SELECCIÓN DE CATEGORÍAS PARA WINE
selected_wine_classes = [5, 6, 7]

wine_selected_df = wine_quality_df[
    wine_quality_df["quality"].isin(selected_wine_classes)
].copy()

print("Muestras originales:", len(wine_quality_df))
print("Muestras con clases 5, 6 y 7:", len(wine_selected_df))

print("\nDistribución:")
print(wine_selected_df["quality"].value_counts().sort_index())

print("\nDistribución porcentual:")
print(
    wine_selected_df["quality"]
    .value_counts(normalize=True)
    .sort_index()
    .mul(100)
    .round(2)
)
```
